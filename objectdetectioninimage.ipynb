```python
import torch

# Load pre-trained YOLOv5 model (YOLOv5s in this case)
model = torch.hub.load('ultralytics/yolov5', 'yolov5s')  # Options: yolov5s, yolov5m, yolov5l, yolov5x
```

```python
# Load an image from a file or URL
img_path = 'path_to_image.jpg'  # Replace with the actual image file path

# Alternatively, you can load an image from a URL
# img_path = '/content/image1.jpg'
```

```python
# Load an image from a file or URL
img_path = '/content/image1.jpg'  # Replace with the actual image file path

# Alternatively, you can load an image from a URL
# img_path = 'image1.jpg' # Make sure the image exists in the current directory or provide the full path

# Perform object detection on the image
results = model(img_path)

# Show the results (this will display the image with bounding boxes and labels)
results.show()
```

```python
# Get results as a pandas DataFrame
detections = results.pandas().xyxy[0]

# Print detection results (bounding boxes, confidence scores, class labels)
print(detections)
```

```python
# Save the result image with detections
results.save(save_dir='output/')  # Replace 'output/' with your desired output folder
```
